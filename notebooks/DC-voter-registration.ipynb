```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DC_voter_registration = pd.read_csv('DC-voter-registration-data-by-year.csv')

DC_voter_registration.columns
```

```python
DC_voter_registration.head()


DC_voter_registration_DsandRs = DC_voter_registration[['Year', 'DEM', 'REP']]

df = DC_voter_registration_DsandRs
```

export for datawrapper

```python
df.to_csv('DC_voter_registration_DsandRs.csv')
```

```python
df['DEM'].unique()
```

Strip out % signs for plotting

```python
df["DEM"] = df["DEM"].str.replace("%", "").astype(float)
df["REP"] = df["REP"].str.replace("%", "").astype(float)
```

```python
plt.figure(figsize=(8, 5))

# Plot the Democratic line
plt.plot(df["Year"], df["DEM"], color='royalblue')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Percent of total registered voters that are Democrats")
plt.title("Registered Democrats in D.C.")

# Set the y-axis range and the y ticks

#Set the x ticks to be the election years, and the limit to show only the data we have, 2012 to 2023
plt.xticks([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
plt.xlim(2014, 2023)


###now I want to overlay with some scatterplot dots to show election years###

#election years
election_years = [2016, 2020]

# filter just those rows
elections = df[df["Year"].isin(election_years)]

# overlay highlighted points
plt.scatter(
    elections["Year"],
    elections["DEM"],
    color="black",
    s=60,        # size of dots
    zorder=3,    # makes sure dots appear above the line
    label="National election year")


###print it

plt.tight_layout()

# Show a legend so you know which line is which
plt.legend()

plt.show()
```

```python
#initial styling

plt.rcParams.update({
    "font.family": "Georgia",
    "font.size": 16
})

#initial styling 

fig, (ax_top, ax_bottom) = plt.subplots(
    2, 1,
    figsize=(9, 5),
    sharex=True,
    gridspec_kw={"height_ratios": [1, 1]}
)
### Make shaded areas on the chart representing each presidency ###

for a in (ax_top, ax_bottom):
    a.axvspan(2009, 2017, color="#dbe9f6", alpha=0.4, label='Democratic President') #Obama shading
    a.axvspan(2017, 2021, color="#f6d6d6", alpha=0.4, label='Republican President') #Trump 1 shading
    a.axvspan(2021, 2025, color="#d6eaf6", alpha=0.4) #Biden shading

### Set up axes + labels ###

# Set the y-axis range and the y ticks
ax_top.set_yticks(
    [70, 75, 80, 85, 90, 95, 100],
    ['70%', '75%', '80%', '85%', '90%', '95%', '100%']
)
ax_top.set_ylim(70,100)

ax_bottom.set_yticks(
    [0, 5, 10, 15, 20, 25,],
    ['0%', '5%', '10%', '15%', '20%', '25%']
)
ax_bottom.set_ylim(0,25)


#Set the x ticks to be the election years and the graph to show only the data we have, 2014 to 2023

ax_bottom.set_xticks([2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
ax_bottom.set_xlim(2014, 2023)

ax_bottom.tick_params(axis="x", labelbottom=True)
ax_top.tick_params(labelbottom=False) # no x labels on the top half of chart

#title and labels on graph

ax_top.set_title('D.C. Voters by Party Registration', loc="left", fontsize=14, fontweight="bold", pad=18)

fig.supylabel(
    "Share of all registered voters (%)", #centering the Y label
    x=0.04,   # adjust left/right position
    fontsize=12
)
# Draw a light grid

ax_top.grid(True, axis="y", color="#e6e6e6", linewidth=1)
ax_bottom.grid(True, axis="y", color="#e6e6e6", linewidth=1)

ax_top.grid(False, axis="x")
ax_bottom.grid(False, axis="x")

# Remove top/right spines

for a in (ax_top, ax_bottom):
    a.grid(True, axis="y", color="#e6e6e6", linewidth=1)
    a.grid(False, axis="x")

    a.spines["top"].set_visible(False)
    a.spines["right"].set_visible(False)
    a.spines["left"].set_color("#cccccc")
    a.spines["bottom"].set_color("#cccccc")

# Ticks styling

ax_top.tick_params(colors="#555555", labelsize=9, bottom=False, labelbottom=False)
ax_bottom.tick_params(colors="#555555", labelsize=9)

plt.setp(ax_top.get_xticklabels(), fontfamily="DejaVu Sans")
plt.setp(ax_bottom.get_xticklabels(), fontfamily="DejaVu Sans")

plt.setp(ax_top.get_yticklabels(), fontfamily="DejaVu Sans")
plt.setp(ax_bottom.get_yticklabels(), fontfamily="DejaVu Sans")

#Sans serif font for ticks

for label in ax_top.get_xticklabels():
    label.set_fontfamily("DejaVu Sans")

for label in ax_bottom.get_xticklabels():
    label.set_fontfamily("DejaVu Sans")


for label in ax_top.get_yticklabels():
    label.set_fontfamily("DejaVu Sans")

for label in ax_bottom.get_yticklabels():
    label.set_fontfamily("DejaVu Sans")


### Draw main data lines on top ###

ax_top.plot(df["Year"], df["DEM"], color="#2b6cb0", linewidth=2.5, label='Voters registered as Democrats')
ax_bottom.plot(df["Year"], df["DEM"], color="#2b6cb0", linewidth=2.5,label='Voters registered as Democrats')

ax_top.plot(df["Year"], df["REP"], color='firebrick', linewidth=2.5,label='Voters registered as Republicans')
ax_bottom.plot(df["Year"], df["REP"], color='firebrick', linewidth=2.5,label='Voters registered as Republicans')


# ### Highlight election years with points ###

# elections = df[df['Year'].isin([2016, 2020])]

# ax_top.vlines(
#     elections["Year"],
#     ymin=73,
#     ymax=80,
#     color="black",
#     linewidth=1
# )

# ax_bottom.vlines(
#     elections["Year"],
#     ymin=2,
#     ymax=9,
#     color="black",
#     linewidth=1
# )

# Creating a broken Y axis

d = 0.015
kwargs = dict(transform=ax_top.transAxes, color='k', clip_on=False)
ax_top.plot((-d, +d), (-d, +d), **kwargs)
ax_top.plot((1 - d, 1 + d), (-d, +d), **kwargs)

kwargs.update(transform=ax_bottom.transAxes)
ax_bottom.plot((-d, +d), (1 - d, 1 + d), **kwargs)
ax_bottom.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)


## make post election years bold ##

for label in ax_bottom.get_xticklabels():
    label.set_fontfamily("DejaVu Sans")

    if label.get_text() in ["2017", "2021"]:
        label.set_fontweight("bold")



## note that 2017 and 2021 begin presidencies ###

ax_bottom.text(
    2017, -0.18,
    "Trump presidency\nbegins",
    ha="center",
    va="top",
    fontsize=8,
    fontfamily="DejaVu Sans",
    transform=ax_bottom.get_xaxis_transform()
)

ax_bottom.text(
    2021, -0.18,
    "Biden presidency\nbegins",
    ha="center",
    va="top",
    fontsize=8,
    fontfamily="DejaVu Sans",
    transform=ax_bottom.get_xaxis_transform()
)



## label each line ##

# Democrat label
ax_top.text(
    2023.1,                        # slightly to the right of last year
    df["DEM"].iloc[-1],            # last y-value
    "Democrats",
    color="#333333",
    fontsize=10,
    fontweight="bold",
    va="center"
)

# Republican label
ax_bottom.text(
    2023.1,
    df["REP"].iloc[-1],
    "Republicans",
    color="#333333",
    fontsize=10,
    fontweight="bold",
    va="center"
)

### overlay a scatterplot to show each year of data ###

#Democratic line 
ax_top.scatter(
    df["Year"],
    df["DEM"],
    s=20,                # point size
    color="#2b6cb0",
    edgecolors="white",  # optional, makes points stand out
    linewidth=0.5,
    zorder=4
)

# Republican line  
ax_bottom.scatter(
    df["Year"],
    df["REP"],
    s=20,
    color="firebrick",
    edgecolors="white",
    linewidth=0.5,
    zorder=4
)

# # Legend

# ax_top.legend(frameon=True, loc="upper right", fontsize=9)

# #choosing simpler non serif font for legend
# leg = ax_top.get_legend()

# for text in leg.get_texts():
#     text.set_fontfamily("DejaVu Sans")
    
plt.tight_layout()
plt.show()
```
