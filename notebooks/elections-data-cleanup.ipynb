```python
import pandas as pd

df = pd.read_csv("ORIGINAL DATA 1976-2024-presidential-elections.csv")
```

data source: MIT Election Data and Science Lab, 2017, "U.S. President 1976–2024", https://doi.org/10.7910/DVN/42MVDX, Harvard Dataverse, V10, UNF:6:xpBppxfswpr+u9xZe7/u7w== [fileUNF]

cut off the old year rows before 2000

```python
df = df.iloc[2411:]
```

Cut out any rows that are for no Republican or Democrat parties

```python
df = df[df["party_detailed"].str.lower().isin(["republican", "democrat"])]
```

Lowercase states

```python
df["state"] = df["state"].str.title()
```

```python
df.head(30)
```

Trim some colums we don't need

```python
#see what's in the office column

df['office'].unique()

df = df.drop(columns=['office','writein','notes','party_simplified','version', 'state_fips', 'state_cen', 'state_ic','state_po','candidate'])
```

Drop DC Columns, and MD and vA

```python
df = df[df["state"] != "District Of Columbia"]

df = df[df["state"] != "Maryland"]

df = df[df["state"] != "Virginia"]
```

Calculate percent to each candidate



```python
df['percent_total_votes'] = df['candidatevotes'] / df['totalvotes']
```

```python
df.head()
```

Need to sort this so I only have one line for each state, with whichever party and candidate received the most votes.

```python
df_winners_only = df.loc[
    df.groupby(['year', 'state'])['candidatevotes'].idxmax()
].reset_index(drop=True)
```

```python
df = df_winners_only
```

Drop a few more columns and isolate each year to save to add as a column to the census migration data tables for each election year.

```python
df = df.drop(columns=['candidatevotes', 'totalvotes'])

df.head()
```

Reorder the columns so state is first

```python
df = df[
    [
        'state',
        'year',
        'party_detailed',
        'percent_total_votes']
]
```

```python
df_2004_election_winners = df[df["year"] == 2004]
df_2008_election_winners = df[df["year"] == 2008]
df_2012_election_winners = df[df["year"] == 2012]
df_2016_election_winners = df[df["year"] == 2016]
df_2020_election_winners = df[df["year"] == 2020]
```
