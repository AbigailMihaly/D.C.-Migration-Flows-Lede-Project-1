This is a notebook to merge each year data from my other notebooks

```python
import pandas as pd
```

Read in all the csv files that I cleaned up from each election year

```python
df_2005 = pd.read_csv('DC-migration-2005.csv')
df_2009 = pd.read_csv('DC-migration-2009.csv')
df_2013 = pd.read_csv('DC-migration-2013.csv')
df_2017 = pd.read_csv('DC-migration-2017.csv')
df_2021 = pd.read_csv('DC-migration-2021.csv')
```

Starting the new dataframe w the 2005 data's migrated adjusted and moe adjusted rows.

```python
df = df_2005[['0', 'migrated_adjusted', 'moe_adjusted']]

df.head()
```

Renaming the states column to be titled 'state names'

```python
df.rename(columns={'0':'state_names'}, inplace=True)

df.head()
```

Rename the columns to be clear they are attatched to the 2005 data before adding in other years.

```python
df.rename(columns={'migrated_adjusted':'migr_adj_2005'}, inplace=True)
df.rename(columns={'moe_adjusted':'moe_adj_2005'}, inplace=True)
```

Add 2009 migrated adjusted and moe adjusted rows.

```python
df['migr_adj_2009'] = df_2009['migrated_adjusted']
df['moe_adj_2009'] = df_2009['moe_adjusted']

df.head()
```

Add 2013 migrated adjusted and moe adjusted rows.

```python
df['migr_adj_2013'] = df_2013['migrated_adjusted']
df['moe_adj_2013'] = df_2013['moe_adjusted']

df.head()
```

Add 2017 migrated adjusted and moe adjusted rows.

```python
df['migr_adj_2017'] = df_2017['migrated_adjusted']
df['moe_adj_2017'] = df_2017['moe_adjusted']

df.head()
```

Add 2021 migrated adjusted and moe adjusted rows.

```python
df['migr_adj_2021'] = df_2021['migrated_adjusted']
df['moe_adj_2021'] = df_2021['moe_adjusted']

df.head()
```

```python
df.to_csv('DC-migration-compiled-election-years.csv')
```
